```python
#Inerhitance for classes and attrbutes
#Helps with code reusability and extensibility
#class child(parent):
class Animal:
    def __init__(self, name):
        self.name = name
        self.is_alive = True

    def eat(self):
        print(f"{self.name} is eating")

    def sleep(self):
        print(f"{self.name} is sleeping")

class Dog(Animal):
    def __init__(self, name):
        self.name = name
        self.is_alive = True

    def eat(self):
        print(f"{self.name} is eating")

    def sleep(self):
        print(f"{self.name} is sleeping")

class Cat(Animal):
    def __init__(self, name):
        self.name = name
        self.is_alive = True

    def eat(self):
        print(f"{self.name} is eating")

    def sleep(self):
        print(f"{self.name} is sleeping")

class Fish(Animal):
    def __init__(self, name):
        self.name = name
        self.is_alive = True

    def eat(self):
        print(f"{self.name} is eating")

    def sleep(self):
        print(f"{self.name} is sleeping")


dog = Dog("Buddy")
cat = Cat("Whiskers")
fish = Fish("Goldie")
print(dog.name)
print(dog.is_alive)
dog.eat()
dog.sleep()
cat.eat()
fish.sleep()
```

```python
#multiple inheritance = inherit from more than one parent class C(A, B)

#multilevel inheritance = inherit from a parents which inherit from another parents

class Prey:
    def flee(self):
        print(f"This animal is fleeing")

class Predator:
    def hunt(self):
        print(f"This animal is hunting")

class Rabbit(Prey):
    pass

class Hawk(Predator):
    pass

class Fish(Prey, Predator):
    pass


rabbit = Rabbit()
hawk = Hawk()
fish = Fish()
rabbit.flee()
hawk.hunt()
```

```python
class Animal:
    def __init__(self, name):
        self.name = name
        
    def eat(self):
        print(f"This {self.name} is eating")
    def sleep(self):
        print(f"This {self.name} is sleeping")

    

class Prey(Animal):
    def flee(self):
        print(f"This {self.name} is fleeing")

class Predator(Animal):
    def hunt(self):
        print(f"This {self.name} is hunting")

class Rabbit(Prey):
    pass

class Hawk(Predator):
    pass

class Fish(Prey, Predator):
    pass


rabbit = Rabbit("Buggy")
hawk = Hawk("Tony")
fish = Fish("Goldie")
rabbit.flee()
hawk.hunt()
rabbit.eat()
fish.sleep()
fish.hunt()
fish.sleep()
```

```python
#create pokemon game with classes and inheritance

class Pokemon:
    def __init__(self, name, level, health):
        self.name = name
        self.level = level
        self.health = health
    def attack(self):
        if self.health > 0:
            print(f"{self.name} is attacking")
        
    def show_info(self):
        print(f"Name: {self.name}\nLevel: {self.level}\nHealth: {self.health}")

    



class FireType(Pokemon):
    def attack(self):
        print(f"{self.name} used Flamethrower!")

class WaterType(Pokemon):
    def attack(self):
        print(f"{self.name} used Water Gun!")

class GrassType(Pokemon):
    def attack(self):
        print(f"{self.name} used Vine Whip!")

class ElectricType(Pokemon):
    def attack(self):
        print(f"{self.name} used Thunder Shock!")


charmander = FireType("Charmander", 5, 50)
squirtle = WaterType("Squirtle", 5, 40)
bulbasaur = GrassType("Bulbasaur", 5, 40)
pikachu = ElectricType("Pikachu", 5, 40)
charmander.show_info()
charmander.attack()
squirtle.show_info()
squirtle.attack()
bulbasaur.show_info()
bulbasaur.attack()
pikachu.show_info()
pikachu.attack()



```

```python
import random

class Warrior:
    def __init__(self, name, hp, hitmin, hitmax):
        self.name = name
        self.hp = hp
        self.hitmin = hitmin
        self.hitmax = hitmax

    def attack(self, energy):
        hit = random.randint(self.hitmin, self.hitmax)
        energy.hp -= hit
        print(f"{self.name} {hit} damage to {energy.name} got {energy.hp} hp left")

class Hero(Warrior):
    def __init__(self, name, hp, hitmin, hitmax, magic):
        super().__init__(name, hp, hitmin, hitmax)
        self.magic = magic
        print(f"{self.name} has {self.magic} magic points")

    def attack(self, energy):
         super().attack(energy)
         self.hp += self.magic
         print(f"{self.name} healed {self.magic} points and now has {self.hp} hp")

    
class Monster(Warrior):
    def __init__(self, name, hp, hitmin, hitmax, rage):
        super().__init__(name, hp, hitmin, hitmax)
        self.rage = rage
        print(f"{self.name} has {self.rage} rage points")

    def attack(self, energy):
         super().attack(energy)
         self.hp += self.rage
         print(f"{self.name} healed {self.rage} points and now has {self.hp} hp")


hero = Hero("Conan", 100, 15, 25, 10)
deadline = Warrior("Deadline", 80, 10, 20)
goblin = Monster("Goblin", 60, 12, 22, 5)
hero.attack(deadline)
deadline.attack(hero)
goblin.attack(hero)
hero.attack(goblin)
deadline.attack(goblin)

```
